# Selección de Peleadores

```python
# Contar las apariciones de cada peleador en las columnas 'r_fighter' y 'b_fighter'
peleas_r = df['r_fighter'].value_counts()
peleas_b = df['b_fighter'].value_counts()

peleas_totales = peleas_r.add(peleas_b, fill_value=0)
peleas_totales
```

Cogemos los peleadores que tengan mas de 25 peleas para que sea mas representativa la muestra

```python
luchadores_mas_25_peleas = peleas_totales[peleas_totales > 25]

print(luchadores_mas_25_peleas)
print(len(luchadores_mas_25_peleas))
```

Voy a coger como métrica de luchador la media de la mediana de sus peleas ganadas y sus medianas de cuando pierde


```python
#Veces que aparecen los peleadores en total
df_filtrado_ganadores = df[df['r_fighter'].isin(luchadores_mas_25_peleas.index)]
print(len(df_filtrado_ganadores))

df_filtrado_perdedores = df[df['b_fighter'].isin(luchadores_mas_25_peleas.index)]
print(len(df_filtrado_perdedores))
```

```python
medianas_ganadores = df_filtrado_ganadores.groupby('r_fighter')[['r_kd', 'r_str', 'r_td', 'r_sub']].median()
print(len(medianas_ganadores))
```

```python
medianas_ganadores.head()
```

```python
medianas_perdedores = df_filtrado_perdedores.groupby('b_fighter')[['b_kd', 'b_str', 'b_td', 'b_sub']].median()
len(medianas_perdedores)
```

```python
medianas_perdedores.head()
```

```python
medianas_ganadores.columns = ['kd', 'str', 'td', 'sub']
medianas_perdedores.columns = ['kd', 'str', 'td', 'sub']
```

```python
metrics_media = (medianas_ganadores + medianas_perdedores) / 2
```

```python
# Unir los datos de los ganadores y perdedores en una sola columna de peleadores
weight_class_ganadores = df_filtrado_ganadores.groupby('r_fighter')['weight_class'].apply(lambda x: x.mode()[0]).reset_index()
weight_class_perdedores = df_filtrado_perdedores.groupby('b_fighter')['weight_class'].apply(lambda x: x.mode()[0]).reset_index()

# Renombrar la columna para facilitar la unión
weight_class_ganadores.rename(columns={'r_fighter': 'fighter', 'weight_class': 'weight_class'}, inplace=True)
weight_class_perdedores.rename(columns={'b_fighter': 'fighter', 'weight_class': 'weight_class'}, inplace=True)

# Combinar ambos grupos de datos (ganadores y perdedores) para obtener la categoría de peso final
weight_classes_combined = pd.concat([weight_class_ganadores, weight_class_perdedores], axis=0)

# Obtener la categoría de peso más frecuente para cada peleador
weight_class_final = weight_classes_combined.groupby('fighter')['weight_class'].apply(lambda x: x.mode()[0])

weight_class_final
```

```python
# Unir la columna de 'weight_class' con 'metrics_media'
metrics_media = metrics_media.merge(weight_class_final, on='fighter', how='left')

# Mostrar el resultado
print(metrics_media.head())
```

```python
def crear_combate(df, index_1, index_2):

    fighter_1 = df.iloc[index_1]
    fighter_2 = df.iloc[index_2]

    print(f"{fighter_1['fighter']} vs {fighter_2['fighter']}")

    kd_diff = fighter_1['kd'] - fighter_2['kd']
    str_diff = fighter_1['str'] - fighter_2['str']
    td_diff = fighter_1['td'] - fighter_2['td']
    sub_diff = fighter_1['sub'] - fighter_2['sub']

    df_enfrentamiento = pd.DataFrame({
        'fighter_1': [fighter_1['fighter']],
        'fighter_2': [fighter_2['fighter']],
        'kd_diff': [kd_diff],
        'str_diff': [str_diff],
        'td_diff': [td_diff],
        'sub_diff': [sub_diff]
    })

    return df_enfrentamiento
```

Haré un sistema de puntuaciones muy sencillo pero podria añadir mas variables sin ningún problema

```python
def predecir_ganador(peleador_1, peleador_2):
    puntos_peleador_1 = 0
    puntos_peleador_2 = 0

    #Knockdowns
    if peleador_1['kd'] > peleador_2['kd']:
        puntos_peleador_1 += 1
    else:
        puntos_peleador_2 += 1

    #Takedowns
    if peleador_1['td'] > peleador_2['td']:
        puntos_peleador_1 += 1
    else:
        puntos_peleador_2 += 1

    #Strikes
    if peleador_1['str'] > peleador_2['str']:
        puntos_peleador_1 += 1
    else:
        puntos_peleador_2 += 1

    #Submisiones
    if peleador_1['sub'] > peleador_2['sub']:
        puntos_peleador_1 += 1
    else:
        puntos_peleador_2 += 1

    #Sistema de puntuaciones
    if puntos_peleador_1 > puntos_peleador_2:
        return peleador_1['fighter']
    elif puntos_peleador_2 > puntos_peleador_1:
        return peleador_2['fighter']
    else:
           # Si tienen los mismos puntos, se desempata por el número de strikes
        if peleador_1['str'] > peleador_2['str']:
            return peleador_1['fighter']
        elif peleador_2['str'] > peleador_1['str']:
            return peleador_2['fighter']
        else:
            return "Empate"  #Tuviesen mismos strikes también, se considera empate.


peleador_1 = metrics_media.iloc[6]  
peleador_2 = metrics_media.iloc[5]  

ganador = predecir_ganador(peleador_1, peleador_2)
print(f"El ganador predicho es: {ganador}")
```
