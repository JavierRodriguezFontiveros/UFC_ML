```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = "../data/raw/pistacho.csv"
pistachos = pd.read_csv(path)
pistachos
```

```python
pistachos["Class"].unique()
```

```python
pistachos.info()
```

```python
clases_pistacho = {'Kirmizi_Pistachio': 0,
                   'Siit_Pistachio': 1}

pistachos['Class'] = pistachos['Class'].map(clases_pistacho)
```

```python
plt.figure(figsize=(14, 12))
sns.heatmap(pistachos.corr(), 
            annot=True,          
            cmap="coolwarm",     
            vmin=-1,
            vmax=1)  

plt.savefig("../data/images/heatmap_pistachos.png", dpi=300, bbox_inches='tight')
```
