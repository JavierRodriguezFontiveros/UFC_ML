# Presentacion de negocio

```python
df= pd.read_csv("../data/processed/data_processed.csv",index_col="Unnamed: 0")
df
```

```python
time_counts = df["time"].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
time_counts_sorted = df['time'].value_counts().sort_values(ascending=False)

explode = [0.1 if i == 0 else 0 for i in range(len(time_counts_sorted))]

colors = plt.cm.tab20.colors[:len(time_counts_sorted)]  

plt.figure(figsize=(8, 8))
time_counts_sorted.plot(kind='pie', startangle=90, explode=explode, colors=colors, autopct=None, labels=None)
plt.ylabel('')
plt.tight_layout();

plt.savefig("../data/images/rounds.png", transparent=True, dpi=300);
```

# Sistema de puntuaciones

```python

medianas_strikes = {'r_str': df['r_str'].mean(),
                     'b_str': df['b_str'].mean()}

medianas_knockdowns = {'r_kd': df['r_kd'].mean(),
                        'b_kd': df['b_kd'].mean()}

medianas_takedowns = {'r_td': df['r_td'].mean(),
                     'b_td': df['b_td'].mean()}

medianas_submissions = {'r_sub': df['r_sub'].mean(),
                         'b_sub': df['b_sub'].mean()}
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))  

# Gráfico para Strikes
axes[0, 0].bar(medianas_strikes.keys(), medianas_strikes.values(), color=['red', 'blue'])
axes[0, 0].set_title('Comparación de Median de Strikes')
axes[0, 0].set_ylabel('Mediana')
axes[0, 0].set_xlabel('Peleador')

#Gráfico para Knockdowns
axes[0, 1].bar(medianas_knockdowns.keys(), medianas_knockdowns.values(), color=['red', 'blue'])
axes[0, 1].set_title('Comparación de Median de Knockdowns')
axes[0, 1].set_ylabel('Mediana')
axes[0, 1].set_xlabel('Peleador')

#Gráfico para Takedowns
axes[1, 0].bar(medianas_takedowns.keys(), medianas_takedowns.values(), color=['red', 'blue'])
axes[1, 0].set_title('Comparación de Median de Takedowns')
axes[1, 0].set_ylabel('Mediana')
axes[1, 0].set_xlabel('Peleador')

#Gráfico para Submissions
axes[1, 1].bar(medianas_submissions.keys(), medianas_submissions.values(), color=['red', 'blue'])
axes[1, 1].set_title('Comparación de Median de Submissions')
axes[1, 1].set_ylabel('Mediana')
axes[1, 1].set_xlabel('Peleador')

plt.tight_layout()


plt.savefig("../data/images/punteo.png", transparent=True, dpi=300);
```
